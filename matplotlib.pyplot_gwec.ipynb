# Explain matplotlib in python

- Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.


- Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.


- pyplot provides various functions for creating different types of plots such as line plots, scatter plots, bar plots, histograms, etc.

Examples include 

- plot(): Line plots

- scatter(): Scatter plots

- bar(): Bar plots

- hist(): Histograms

- boxplot(): Box plots

- pie(): Pie charts

- imshow(): Image plots


Customization:

- pyplot allows you to customize various aspects of your plot, including titles, labels, colors, linestyles, markers, etc.


- Functions like plt.title(), plt.xlabel(), plt.ylabel(), plt.legend(), plt.grid(), plt.xlim(), plt.ylim(), etc., are used for customization.

```python
import matplotlib.pyplot as plt
```

```python
#simple line chart

x=[1,2,4,5,6]
y=[4,3,4,5,6]
plt.plot(x,y,color="red",marker="o",markersize=3,linewidth=1)
plt.title("simple line chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
x=[23,4,5,6,7]
y=[1,32,4,5,6]
z=[4,4,5,6,7]
plt.plot(x,y,label="line one")
plt.plot(y,z,label="second line")
plt.title("multiple line chart")
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
x=np.linspace(0,10,200)
y=np.sin(x,y)
plt.plot(x,y,color='blue',linestyle="--",linewidth="2",label="sin(x),markersize=4")
plt.title("customised line chart")
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.legend()

plt.grid(True)
plt.show()
```

```python
#bar plot

name=["ajit","aniket","nisha","neha"]
age=[24,25,21,24]
plt.bar(name,age,color="red",width=0.25)
plt.title("simple names")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
name=["ajit","aniket","nisha","neha"]
age=[24,25,21,24]
plt.barh(name,age,color="red")
plt.title("simple names")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
#scatter plot
x=[3,4,5,6,7,34,5,5]
y=[5,6,7,8,8,3,5,7]
plt.scatter(x,y,color="red",alpha=0.35)
plt.title("simple scatter plot")
plt.xlabel("x-axis ")
plt.ylabel("y-axis ")
plt.show()
```

```python
name=["ajit","aniket","nisha","neha"]
age=[24,25,21,24]
plt.hist(name,color="red",bins=30)
plt.title("simple names")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
data =np.random.randn(1000)
plt.hist(data,bins=30,color="skyblue")
plt.title("HISTOGRAM")
plt.xlabel("values")
plt.ylabel("data")
plt.show()
```

```python
age=[100,23,5,4,1]
name=["fridge","chair","cup","glass","bottle"]
plt.pie(age,labels=name,autopct="%1.1f%%")
plt.title("simple pie chart")
plt.axis("equal")
plt.show()
```
